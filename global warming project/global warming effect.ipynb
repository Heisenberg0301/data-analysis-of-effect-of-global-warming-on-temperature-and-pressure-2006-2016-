```python
def reshape(lst, n):
    return [lst[i*n:(i+1)*n] for i in range(len(lst)//n)]
#print(january)
jan=reshape(january,744)
feb=reshape(february2,672)
mar=reshape(march,744)
apr=reshape(april,720)
ma=reshape(may,744)
jun=reshape(june,720)
jul=reshape(july,720)
aug=reshape(august,744)
sep=reshape(september,720)
oct=reshape(october,744)
nov=reshape(november,720)
dec=reshape(december,744)
#print(len(january))

print(len(feb[0]))
```

```python
apr_temp=[]
apr_hum=[]

def avg(apr_temp,apr_hum,apr):
    for i in apr:
        temp_sum=0
        hum_sum=0
        for j,k in i:
            hum_sum+=j
            temp_sum+=k
        
        apr_temp.append(temp_sum/len(i))
        apr_hum.append(hum_sum/len(i))
avg(apr_temp,apr_hum,apr)   
print(apr_temp)
print(apr_hum)
```
